```python
y = df_maths['finalResult']
```

```python
X = df_maths.drop(['finalResult', 'academicGrade', 'G3'], axis = 1)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = split_data(X,y)
```

```python
model = KNeighborsClassifier(n_neighbors= 4) ## Cette ligne change si vous changez d'algo.
```

```python
knn = create_model(model= model, X_train = X_train)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
y_pred
```

```python
dico = {"pass":1, "fail":0}
_true = y_test.apply(lambda x: 1 if x== "pass" else 0)
_pred  = []
for col in y_pred:
    if col == 'pass':
        _pred.append(1)
    else:
        _pred.append(0)
```

```python
precision_score(y_pred= _pred, y_true = _true)
```

```python
recall_score(y_pred= _pred, y_true = _true)
```

```python
accuracy_score(y_pred= _pred, y_true = _true)
```

```python
confusion_matrix(y_pred= _pred, y_true = _true)
```
